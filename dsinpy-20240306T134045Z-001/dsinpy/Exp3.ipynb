```python
import pandas as pd

bank_dataset = pd.read_csv(r".\bank_2.csv")
print(bank_dataset)
```

```python
bank_dataset['education'].replace({'primary':'0','secondary':'1','tertiary':'2','unknown':' 3'}, inplace = True)
print(bank_dataset.head())
```

```python
bank_dataset['marital'].replace({'married':'1','single':'0'}, inplace = True)
print(bank_dataset.head())
```

```python
bank_dataset['housing'].replace({'yes':'1','no':'0'}, inplace = True)
print(bank_dataset.head())
```

```python
# bank_dataset['loan'].replace({'yes':'1','no':'0'}, inplace = True)
print(bank_dataset.head())
```

```python
bank_dataset['balance']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


bank_dataset_copy = bank_dataset
sns.displot(bank_dataset_copy["balance"])
```

```python
for x in bank_dataset_copy["balance"]:
    print(x)
```

```python
# finding IQR
percentile25 = bank_dataset_copy['balance'].quantile(0.25)
percentile75 = bank_dataset_copy['balance'].quantile(0.75)
iqr = percentile75 - percentile25
```

```python
# finding upper and lower limit
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
# finding outliers using iqr
bank_dataset_copy[bank_dataset_copy['balance'] > upper_limit]
bank_dataset_copy[bank_dataset_copy['balance'] < lower_limit]
```

```python
# trimming outliers
new_df = bank_dataset_copy[bank_dataset_copy['balance'] < upper_limit]
new_df.shape
```

```python
# compare plots after trimming
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(bank_dataset['balance'])
plt.subplot(2,2,2)
sns.boxplot(bank_dataset_copy['balance'])
plt.subplot(2,2,3)
sns.distplot(new_df['balance'])
plt.subplot(2,2,4)
sns.boxplot(new_df['balance'])
plt.show()
```
